# Picewise-Linear Interpolation

### Generate 10 points on the plane, with increasing values of x-coordinates

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 10)
y = np.random.uniform(0, 10, 10)
```

### Draw the Piecewise Linear Interpolating function for this Data, along with the Data points

```python
plt.plot(x, y, color='red')
plt.scatter(x, y, color='blue', zorder=999)
plt.title('Piecewise Linear Interpolating Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

### Draw the Not-a-Knot Cubic Spline Interpolating this Data on the same graph as the PLI function

```python
from scipy.interpolate import CubicSpline
cub_spl = CubicSpline(x, y)
xx = np.linspace(0, 10, 100)
yy = cub_spl(xx)
plt.title('Not-a-Knot Cubic Spline Interpolation')
plt.plot(xx, yy, color='yellow')
plt.plot(x, y, color='red')
plt.scatter(x, y, color='blue', zorder=999)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
