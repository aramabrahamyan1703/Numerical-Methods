### We consider the following BVP for a 2nd order ODE:

$$\left\{\begin{split}
y^{\prime\prime}-4y=x,\  \ x∈[0,2] \\
y(0) = 0,\ y(2)=1\\
\end{split}\right.$$

It is known that the true solution is given by:
$$y=\frac{1}{4−4e^8}\cdot[6e^{4−2x} −6e^{4+2x} −x+e^{8}x]$$

Solve this BVP for the mesh size equal to 100 using the Tridiagonal System
direct solution method, plot the graph of the numerical solution and the true solution

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x): # actual function
    return 1/(4-4*np.exp(8))*(6*np.exp(4-2*x)-6*np.exp(4+2*x)-x+np.exp(8)*x)
n = 100 # mesh size
a = 0 #interval left endpoint
b = 2 #interval rigth endpoint
h = (b-a)/n #step-size
x = np.linspace(a,b,n-1)
```

```python
A = np.zeros((n-1, n-1))
c = np.zeros(n-1)
for i in range(1, n-1):
    A[i][i] = (-4*h**2-2)
    A[i-1][i] = 1
    A[i][i-1] = 1
for i in range(n-1):
    c[i] = a + i * h**3
c[-1] -= 1
y=np.linalg.solve(A,c)
plt.plot(x,f(x), label='True Solution')
plt.plot(x, y)
plt.legend()
```

Solve the BVP for the mesh size equal to 100 using one of the iterative methods. Compare, graphically, with the above numerical solution and the true
solution.

```python
z = np.zeros(n-1)
z[-1] = 1
for j in range(4000):
     for i in range(1, n-2):
         z[i] = (x[i+1] * h**2 -z[i+1] - z[i-1])/(-2-4*h**2)
plt.plot(x,f(x), label='True Solution')
plt.plot(x, y, label='Direct Solution')
plt.plot(x, z, label='Iterative Solution')
plt.legend()
```
