Assume we are given the following Data:
$$x: −4,−3,−2,0,4,5\\
y: 3,6,1,−2,3,1$$
a. Construct the Interpolating Polynomial for this Dataset;

```python
from scipy.interpolate import lagrange
import numpy as np
import matplotlib.pyplot as plt

x = np.array([-4, -3, -2, 0, 4, 5])
y = np.array([3, 6, 1, -2, 3, 1])
poly = lagrange(x, y)
print(np.poly1d(poly))
```

$$\text{b. Plot the graph of the IP over some interval,\\
covering all points, and also add the\\
Interpolation Points on that graph.}$$

```python
xx = np.linspace(-4, 5, 100)
yy = np.polyval(poly, xx)
plt.plot(xx, yy, color='red', label='Interpolation')
plt.scatter(x, y, label='Data points', zorder=999)
plt.legend()
plt.show()
```
