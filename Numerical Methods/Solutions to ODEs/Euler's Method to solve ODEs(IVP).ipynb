### Here we consider the following IVP for a 1st order ODE:

$$\left\{\begin{split}
y^{\prime}=y+x^2,\  x∈[0,2] \\
y(0) = 1\\
\end{split}\right.$$
• Check that the following is a solution to this IVP:
\begin{equation*}
y(x) = 3e^x − x^2 − 2x − 2
\end{equation*}
Soultion:
$y(0)=3e^0 - 0^2 -0-2=3-2=1 \checkmark$
$y^{\prime}=3e^x - 2x - 2\\
y+x^2 = 3e^x - x^2+x^2-2x-2=3e^x - 2x-2 = y{\prime} \checkmark\\
\implies y(x) = 3e^x − x^2 − 2x − 2 is\ solution\ to\ this\ IVP$


• Using the Euler’s method, solve this IVP with different values of the mesh size. Give the graphs of your numerical solution and the true solution on the same frame.

• Print also the maximum absolute error for each mesh size.

```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2, figsize=(20,5))
def true_sol(x):
    return 3* np.exp(x) - x**2 - 2*x -2
a = 0
b = 2
val = {}
for n in [10, 20, 50, 100]:
    h = (b - a)/n
    x = np.linspace(a, b, n+1)
    y = np.zeros(n+1)
    y[0] = 1 
    for i in range(n):
        y[i+1] = y[i] + h * (y[i] + (a+i*h)**2)
    ax[0].plot(x, y, label = f'$Approximate\ Solution n =$ {n}')
    ax[1].plot(x, true_sol(x) - y, label = f'$Absolute\ Error\ for\ n =$ {n}')
    val[f'Maximum Absolute Error for n = {n}'] = np.max(abs(true_sol(x) - y))
ax[0].plot(x, true_sol(x), label = '$True\ solution$')
ax[0].legend()
ax[1].legend()
ax[0].set_title('True & Approximate solution')
ax[1].set_title('Absolute Error')
val
```
