###### Draw the graphs of $f$ and the Taylor’s Polynomial constructed above in $[−2\pi, 2\pi]$, on the same figure.

```python
import sympy as sp
```

```python
x = sp.Symbol('x')
f = sp.sin(2*x)
tpoly = f.subs(x, 0)
f_diff = sp.diff(f,x)
deg = 5
for i in range(deg):
    tpoly += (f_diff.subs(x,0))/(sp.factorial(i+1)) * x**(i+1)
    f_diff = sp.diff(f_diff,x)
tpoly
```

```python
p1 = sp.plot(tpoly, (x, -2*sp.pi, 2*sp.pi), ylim=[-7, 7], xlabel = 'x', ylabel = 'y', show=False, legend = True, label = 'Taylor\'s polynomial')
p2 = sp.plot(f, (x, -2*sp.pi, 2*sp.pi), ylim=[-7, 7], xlabel = 'x', ylabel = 'y', show=False, legend = True, label = '$sin(2x)$')
p1.extend(p2)
p1.show()
```
