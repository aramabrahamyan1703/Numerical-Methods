$$\text{Let} f(x) = arctg(x), x ∈ R, \text{and let the following nodes are given  }\\
x : −2, −1, −0.5, 0, 0.3, 0.7, 1, 1.5, 2, 3.\\
\text{a. Obtain the Interpolating Polynomial } P(x)\\
\text{ of the corresponding order for } f(x) \text{ at the nodes }x;$$






```python
import numpy as np
from scipy.interpolate import lagrange
from matplotlib import pyplot as plt

nodes = np.array([-2, -1, -0.5, 0, 0.3, 0.7, 1, 1.5, 2, 3])
y = np.arctan(nodes)
poly = lagrange(nodes, y)
print(np.poly1d(poly))
```

$$\text{b. Calculate the approximation error of }f \text{ by }P \text{ at the points }\\
x = 0.3, x = 0.9, x = 2.8;\\$$

```python
err1 = abs(np.arctan(0.3)-np.polyval(poly, 0.3))
err2 = abs(np.arctan(0.9)-np.polyval(poly, 0.9))
err3 = abs(np.arctan(2.8)-np.polyval(poly, 2.8))
print(f"The error at x = 0.3 is {err1},\nThe error at x = 0.9 is {err2},\nThe error at x = 2.8 is {err3}")
```

$$\text{c. Plot the interpolation nodes with red asterisks, the graph of } f(x) \text{ and } P(x)\\
 \text{ for } x ∈ [−2, 3] \text{ on the same graphical window};\\$$

```python
xx = np.linspace(-2, 3, 100)
yy = np.arctan(xx)
zz = poly(xx)
plt.plot(xx, yy, color='blue', label='arctan(x)')
plt.plot(xx, zz, color='green', label='Polynomial')
plt.scatter(nodes, y, color='red', marker='*', zorder=999)
plt.legend()
plt.show()
```

$$\text{d. Plot on a new graphical window the Absolute Error function }\\
|f(x) − P(x)| \text{ for } x ∈ [−2, 3];\\$$

```python
plt.plot(xx, abs(yy-zz), label='Absolute Error Function')
plt.legend()
plt.show()
```

$$\text{e. Find the maximal error }|f(x) − P(x)|\\
\text{by calculating the maximal element of the vector }|f(x) − P(x)|,\\
\text{where }x \text{ runs from }−2 \text{ to }3 \text{ with the step size }0.01.$$

```python
xx = np.arange(-2, 3, 0.01)
zz = abs(np.arctan(xx) - np.polyval(poly, xx))
print(f"The maximal Error is {np.max(zz)}")
```
